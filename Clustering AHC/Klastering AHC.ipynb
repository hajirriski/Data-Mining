```python
#import library
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
```

```python
x = np.array([[1,2],[2.5,4.5],[2,2],[4,1.5],[4,2.5]])
```

```python
print(x)
```

```python
Z = linkage(x)
```

```python
print(Z)
```

```python
#plot dendogram
plt.figure(figsize=(25, 10))
plt.title('Dendogram Hierarchical Klastering')
plt.xlabel('data')
plt.ylabel('jarak')
dendrogram(Z)
plt.show()
```
